```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ch = '#FF7F0E'
cm = '#1F77B4'
```

```python
df = pd.read_csv('2020.02.csv')
df
```

```python
df.columns.tolist()
```

```python
df = df[df['Estoy trabajando en'] == 'Argentina']
df = df[df['Tengo'] < 100]
df[['Me identifico', 'Tengo', '¿Cómo venís llevando la cuarentena?', '¿Tenés hijos/as menores de edad?', '¿Con cuántas personas estás conviviendo?', '¿Con quiénes convivís?','¿Tenés un espacio dedicado para el trabajo?',
 '¿Tenés que compartir tu equipo de trabajo con alguien?',
 '¿Qué tipo de cuarentena estás haciendo?',
 '¿Cambió tu situación laboral a raíz de la pandemia?',
 '¿Qué tanto sentís que te está apoyando tu empresa en esta situación?',
 '¿Cómo se vio afectada tu empresa?',
 '¿Instauraron algún beneficio nuevo?',]]
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot(1,1,1)
ax.bar(df[df['Me identifico'] == 'Hombre']['Tengo'].value_counts().index-0.2, df[df['Me identifico'] == 'Hombre']['Tengo'].value_counts().values, width=0.4, color=ch, label='Hombre')
ax.bar(0, 0, label='Mujer')
ax2 = ax.twinx()
ax2.bar(df[df['Me identifico'] == 'Mujer']['Tengo'].value_counts().index+0.2, df[df['Me identifico'] == 'Mujer']['Tengo'].value_counts().values, width=0.4, color=cm, label='Mujer')
ax.legend()
ax.set_ylabel('Hombres')
ax2.set_ylabel('Mujeres')
ax.set_xlabel('Edad')
ax.set_xlim((df['Tengo'].min(), df['Tengo'].max()))
None
```

```python
col = '¿Cómo venís llevando la cuarentena?'

plt.figure(figsize=(10, 10))
ax = plt.subplot(1,1,1)
ax.set_title(f'{col} por género')
ax.bar(df[df['Me identifico'] == 'Hombre'][col].value_counts().index-0.2, df[df['Me identifico'] == 'Hombre'][col].value_counts().values, width=0.4, color=ch, label='Hombre')
ax.bar(0, 0, label='Mujer')
ax2 = ax.twinx()
ax2.bar(df[df['Me identifico'] == 'Mujer'][col].value_counts().index+0.2, df[df['Me identifico'] == 'Mujer'][col].value_counts().values, width=0.4, color=cm, label='Mujer')
ax.legend()
ax.set_ylabel('Hombres')
ax2.set_ylabel('Mujeres')
ax.set_xlabel(col)
None
```

```python
col = '¿Cómo venís llevando la cuarentena?'
col2 = '¿Con cuántas personas estás conviviendo?'
val2 = 0

plt.figure(figsize=(10, 10))
ax = plt.subplot(1,1,1)
ax.set_title(f'{col} por vivir con gente')
ax.bar(df[df[col2] == val2][col].value_counts().index-0.2, df[df[col2] == val2][col].value_counts().values, width=0.4, color=ch, label='Sin gente')
ax.bar(0, 0, label='Con gente')
ax2 = ax.twinx()
ax2.bar(df[df[col2] != val2][col].value_counts().index+0.2, df[df[col2] != val2][col].value_counts().values, width=0.4, color=cm, label='Con gente')
ax.legend()
ax.set_ylabel('Sin gente')
ax2.set_ylabel('Con gente')
ax.set_xlabel(col)
None
```

```python
col = '¿Con quiénes convivís?'
col2 = '¿Cómo venís llevando la cuarentena?'
col3 = 'Me identifico'
val3 = 'Hombre'
v = df[col].value_counts()
v = v[v > 10]
plt.figure(figsize=(30, 20))
plt.suptitle(f'{col2} según {col}')
assert(v.shape[0] == 15)
for i, val in enumerate(v.index.tolist()):
    ax = plt.subplot(3, 5, i+1)
    ax.bar(df[(df[col] == val) & (df[col3] == val3)][col2].value_counts().index-0.2, df[(df[col] == val) & (df[col3] == val3)][col2].value_counts().values, width=0.4, color=ch, label='Hombres')
    ax.bar(0, 0, color=cm, label='Mujeres')
    ax.twinx().bar(df[(df[col] == val) & (df[col3] != val3)][col2].value_counts().index+0.2, df[(df[col] == val) & (df[col3] != val3)][col2].value_counts().values, width=0.4, color=cm, label='Mujeres')
    ax.set_xlim((0, 5))
    ax.set_xticks([1,2,3,4])
    ax.set_title(val)
    ax.legend(loc=2)
```

```python
col = '¿Qué tipo de cuarentena estás haciendo?'
col2 = '¿Cómo venís llevando la cuarentena?'
col3 = 'Me identifico'
val3 = 'Hombre'
v = df[col].value_counts()
v = v[v > 10]
plt.figure(figsize=(12, 12))
plt.suptitle(f'{col2} según {col}')
assert(v.shape[0] == 4)
for i, val in enumerate(v.index.tolist()):
    ax = plt.subplot(2, 2, i+1)
    ax.bar(df[(df[col] == val) & (df[col3] == val3)][col2].value_counts().index-0.2, df[(df[col] == val) & (df[col3] == val3)][col2].value_counts().values, width=0.4, color=ch, label='Hombres')
    ax.bar(0, 0, color=cm, label='Mujeres')
    ax.twinx().bar(df[(df[col] == val) & (df[col3] != val3)][col2].value_counts().index+0.2, df[(df[col] == val) & (df[col3] != val3)][col2].value_counts().values, width=0.4, color=cm, label='Mujeres')
    ax.set_xlim((0, 5))
    ax.set_xticks([1,2,3,4])
    ax.set_title(f'{col}: {val}')
    ax.legend(loc=2)
```
